```python
# extract [image name + label] from score excel and save data as csv file
import pandas as pd
file_path = '/home/eiao/pythoncode/FaceReco/datasets/labels.xlsx'
df = pd.read_excel(file_path)
score_df = df['score'].dropna()
score = score_df.tolist()
label = []
for i in range(1100):
    label.append(['{0:04}'.format(i+1) + '.jpg'] + [score[3*i: 3*i+3]])
label_df = pd.DataFrame(label)

label_df.to_csv('/home/eiao/pythoncode/FaceReco/datasets/data.csv', index=False, header=['image', 'labels'])
```
